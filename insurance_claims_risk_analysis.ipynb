# Insurance Claims Analysis: Identifying Risk Drivers and Cost Segments
An exploratory and decision-focused analysis of insurance claim frequency and severity


### 1. Problem Context & Objective

Insurance providers must balance risk, pricing, and customer retention.
Understanding which customer and policy characteristics are associated with higher claim costs is critical for underwriting, pricing, and operational decision-making.

This analysis aims to:
- Identify patterns in insurance claim frequency and severity
- Segment policies by risk level
- Highlight characteristics associated with higher expected claim costs

The goal is not to build a predictive model, but to extract interpretable insights that could support real-world insurance decisions.


### 2. Dataset Overview

Source:
- Dataset: Auto Insurance Claims Data
- Origin: Kaggle (public dataset)

Structure:
- Each row represents: [policyholder / policy-year / claim record]
- Target variables: [to be identified]
- Feature groups:
  - Customer attributes
